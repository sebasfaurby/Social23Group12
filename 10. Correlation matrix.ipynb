```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('master_df.csv', sep=';')
```

```python
df.drop(['Unnamed: 0','date','Ticker'], axis=1, inplace=True)
df
```

```python
# Correlation matrix using pandas
corrMatrix = df.corr()
```

```python
sn.heatmap(corrMatrix, annot=True, annot_kws={"size": 7})
# change the font size
# sn.set(font_scale=1)
plt.show()
```

```python
df_market_cap = df[df['Market Cap'] < 100000000]
corrMatrix2 = df_market_cap.corr()
```
